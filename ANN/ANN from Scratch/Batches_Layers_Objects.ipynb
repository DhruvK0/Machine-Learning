```python
import numpy as np

inputs = [[1, 2, 3, 2.5],
         [2.0, 5.0, -1.0, 2.0],
         [-1.5, 2.7, 3.3, -0.8]]

weights = [[0.2, 0.8, -0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]
```

```python
#Transpose a matrix when the shape is not right for matrix multiplication
#Transpose means to make the columns the rows, and the rows the columns

transpose_weights = np.array(weights).T
print(transpose_weights)
```

# Transposing a Matrix


```python
output = np.dot(inputs, transpose_weights) + biases
print(output)
```

# 2 Layer weights and biases

```python
inputs = [[1, 2, 3, 2.5],
         [2.0, 5.0, -1.0, 2.0],
         [-1.5, 2.7, 3.3, -0.8]]

weights = [[0.2, 0.8, -0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

#layer 2 weights and biases
weights2 = [[0.1, -0.14, 0.5],
            [-0.5, 0.12, -0.33],
            [-0.44, 0.73, -0.13]]

biases2 = [-1, 2, -0.5]

layer1_output = np.dot(inputs, transpose_weights) + biases
layer2_output = np.dot(layer1_output, np.array(weights2).T) + biases2

print(layer2_output)
```

# Forward Propagation

```python
np.random.seed(0)

X = [[1, 2, 3, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]]

class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.10 * np.random.randn(n_inputs, n_neurons) #multiply by 0.10 so limit the range of weights from -0.1 to 0.1
        self.biases = np.zeros((1, n_neurons))
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases

        
layer1 = Layer_Dense(4, 5)
layer2 = Layer_Dense(5, 2)

#Forward Propagation Example
layer1.forward(X)
print(layer1.output)

layer2.forward(layer1.output)
print(layer2.output)
```
