```python
#Divide Wine quality into two lassifications: bad and good --- *Note: try and add a 3rd classification medium
raw_data['quality'] = pd.cut(raw_data['quality'], bins = (2,6.5,8), labels = ['bad', 'good'])
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
label_quality = LabelEncoder()
```

```python
raw_data['quality'] = label_quality.fit_transform(raw_data['quality'])
```

```python
sns.countplot(raw_data['quality'])
```

```python
x = raw_data.drop('quality', axis=1)
y = raw_data['quality']
y
```

```python
#Split training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42 )
```

```python
#Scale features for a more accurate model
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.svm import SVR
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
svc = SVC()
svc.fit(x_train,y_train)
```

```python
pred_svc = svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, pred_svc))
```

```python
#Update SVC with optimized parameters using rbf kernel
svc_fin = SVC(C = 1.2, gamma = 0.9, kernel='rbf')
svc_fin.fit(x_train,y_train)
pred_svc_fin = svc_fin.predict(x_test)
print(classification_report(y_test, pred_svc_fin))
```
