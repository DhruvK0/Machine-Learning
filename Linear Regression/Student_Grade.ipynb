```python
## Create new data frame with selected features
attributes = ['school','sex','studytime','internet','Walc','romantic','goout','address','G1','G2','G3']
data_fin = pd.DataFrame(data_1,columns=attributes)
data_fin.head()
```

```python
## Use dummy variables for regression analysis
data_fin['school'] = data_fin['school'].map({'GP': 0,'MS': 1})
data_fin['sex'] = data_fin['sex'].map({'M': 1,'F': 0})
data_fin['internet'] = data_fin['internet'].map({'no': 0,'yes': 1})
data_fin['romantic'] = data_fin['romantic'].map({'no': 0,'yes': 1})
data_fin['address'] = data_fin['address'].map({'U': 0,'R': 1})
```

```python
data_fin.describe(include='all')
```

```python
data_fin_corr = data_fin.corr()
sns.heatmap(data_fin_corr)
```

```python
##Create train and test split data
from sklearn.model_selection import train_test_split
x = data_fin[['school','sex','studytime','internet','Walc','romantic','goout','address','G1','G2']]
y = data_fin['G3']
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state= 51)
```

```python
print(" Shape of x_train = ", x_train.shape)
print(" Shape of x_test = ", x_test.shape)
print(" Shape of y_train = ", y_train.shape)
print(" Shape of y_test = ", y_test.shape)
```

```python
## Perform feature scaling before regression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#Fit the data to the regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(x_test)
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted Values'])
y_test_df = pd.DataFrame(np.array(y_test), columns=['Actual Values'])
pd.concat([y_pred_df , y_test_df], axis=1)
```

```python
reg.score(x_test,y_test)
```

```python
# Let's use the handy function we created
def adj_r2(x,y):
    r2 = reg.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adj_r2(x_train,y_train)
```
